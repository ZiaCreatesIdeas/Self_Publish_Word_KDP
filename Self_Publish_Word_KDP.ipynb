Welcome to this notebook. The purpose is to incrementally document the steps required for converting a Word document to KDP ready PDF.

The first step is to determine the size of the book we would like. Everything else flows from there. We have chosen 5.5 x 8.5".

For this approach, we will have to use Microsoft Word for Desktop. The cloud version does not support 'mirror margin' and we need that feature to address the alternating odd and even pages.

Also note that there are points where it is critical to be 'print layout' mode. This should be selected at the outset of the project and more or less be our default.
